## Hadoop Distributed File System

<b>HDFS</b> is the primary storage system of hadoop which stores the data files in distributed manner on commodity hardware.

- NameNode
- DataNode

### MapReduce

<b>MapReduce</b> is the core Hadoop ecosystem compinent which provides data processing.
<br>
MapReduce words in two phases:
- Map phase
- Redice phase

### YARN

<b>Hadoop YARN</b> is a Hadoop ecosystem component that provides the resources management.

### Hive


Open source data warehouse system for querying and analyzing large datasets stored in Hadoop files.

Hive use language called HiveQL (HQL), which is similar to SQL.
- Metastore
- Driver
- Query compiler
- Hive server

### Pig    

<b>Apache Pig</b> is a tool for analyzing and querying data files that stored in HDFS.

Pig uses <b>PigLatin</b> Language which is very similar to SQL

### HBase

<b>Apache HBase</b> is a distributed database that is designed to store structured data in tables that could have billions of rows and millions of columns.

It is highly scalable, distributed, and NoSQL database that is built on top of HDFS.

Provide real-time access to read or write data in HDFS

### Spark

An open-source distributed computing engine for processing and analyzing vast volumes of real-time data.

Spark provides an in-memory computation of data, used to process and analyze real-time streaming data such as Instagram, Twitter feeds and banking data, any many other things.


### Apache Sqoop

<b>Sqoop</b> imports data from external sources into related Hadoop ecosystem components like HDFS, Hbase or Hive.

Sqoop works with RDBMS such as teradata, Netezza, oracle, MySQL

### Apache Flume

<b>Flume</b> efficiently collects, aggregate and populate a large amount of data from its source and write it to HDFS

Using Flume, we can get the data from multiple servers immediately into Hadoop

### Ambari

Ambari is a management platform for provisioning, managing, monitoring and securing apache Hadoop cluster.

provide consistent, secure platform for operational control

### Zookeeper

<b>Apache Zookeeper</b> is a centralized service for maintaining configuration information, naming, providing distributed synchronization, and providing group services.

### Oozie

It is a wordflow scheduler system for managing apache Hadoop jobs.

Oozie combines multiple jobs sequentially into one logical unit of word.
