# Hadoop Introduction

- Hadoop is an Apache's open source platform written in java which allows distributed procrssing of large datasets across cluster of commodity hardware using simple programming models.
  
- It was originated by Doug Cutting and Mike Cafarella.

  
- Hadoop is designed to *horozontally* scale up from single machine to thousands of machines, each serves computation and storage purpose.

### Hadoop Architecture

Hadoop contains three core components:
- <b>Hadoop Distributed File System (HDFS)</b> - Storage layer

- <b>Map-Reducer</b> - Process layer

- <b>YARN</b> - Resource management layer




### 1. Hadoop Distributed File System (HDFS)

- HDFS has a master-slave architecture.

- The data files are divided into multiple blocks

- Hadoop stores these blocks in a distributed manner on the cluster of slave nodes. On the master, we have all the metadata stored.

#### HDFS has two daemons
1. NameNode
2. DataNode

### 2. MapReduce

It is thedata preprocessing layer of Hadoop. It processes data in two phases.

1. Map Phase
2. Reduce Phase

### 3. YARN

YARN is short for Yet Another Resource Negotiator. It has the fillowing components:-

1. Resource Manager
2. Node Manager
