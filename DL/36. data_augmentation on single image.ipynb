```python
from tensorflow.keras.utils import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = load_img('data_augmentation/cat.jpg', target_size=(200,200))
```

```python
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
)
```

```python
img = img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1, 200, 200, 3)
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir='data_augmentation/output'):
    i = i + 1

    if i == 10:
        break
```
