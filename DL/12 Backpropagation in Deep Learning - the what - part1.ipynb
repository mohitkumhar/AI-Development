# Backpropagation

- It is the Algorithm to train the Neural Network
- OR it find the optimimal value of weight and bias to which our neural network gives best prediction 

### Pre requisite
- Gradient Descent
- forward Propogation

- in backpropagation we basically have to update the value of weights and bias (in searching of its best value)
- we use Gradient Descent algo to find the better value of weights and bias
- and this formula is used to update the values of weights
  

   - W = Weights
   - alpha = Learning rate
   - J = Cost
