### **Pooling in CNN:**  
Pooling is a technique used to reduce the size of an image while keeping the most important information. It helps in making the model faster and reducing overfitting.  

#### **Types of Pooling:**  
1. **Max Pooling:** Takes the highest (maximum) value from a region of the image.  
2. **Average Pooling:** Takes the average of all values in a region.  

Pooling helps in keeping important features while making the image smaller.

```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
model = Sequential()

model.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid'))
model.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid'))

model.add(Flatten())

model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```
