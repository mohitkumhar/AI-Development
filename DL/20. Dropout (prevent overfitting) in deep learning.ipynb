## Dropout in Deep Learning
- Dropout is a regularization technique used in deep learning to prevent overfitting. Overfitting happens when a model learns patterns specific to the training data but fails to generalize to new data.

## How Dropout Works
- During training, dropout randomly deactivates (drops) a fraction of neurons in a layer.
- This means some neurons are ignored in each training step, making the network rely on different neurons each time.
- As a result, the model does not become too dependent on specific neurons and learns more generalized patterns.

```python
from tensorflow.keras.layers import Dropout

# Adding dropout to a layer
layer = Dropout(0.5)  # 50% of neurons will be dropped during training
```
