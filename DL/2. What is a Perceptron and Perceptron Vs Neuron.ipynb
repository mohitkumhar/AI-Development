### What is a Perceptron?
It is an Algorithm, used for supervised machine learning


- we have to find the value of w1, w2 and b by training
- the onput fields are X1, X2 where W1 and W2 are the weight of that inputs
- weight shows us the feature importance, means high weight means more important input
- it is used for a linear data and it fails for a non linear data
- 

#### Gometric Intution
Z = W1 X1 + W2 X2 + b

<b>Where</b>
<pre>
    
             ___
            |  1   Z >= 0
y = f(Z) =  |  
            |  0   Z < 0

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
sns.scatterplot(x='cgpa', y='resume_score', data=df, hue='placed')
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import Perceptron
```

```python
pe = Perceptron()
```

```python
pe.fit(X_train, y_train)
```

```python
pe.coef_  # these are the values of W1 and W2 of input data in perceptron
```

```python
pe.intercept_   # this is the value of b
```

### finding the accuracy

```python
y_pred = pe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred, y_test)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=pe, legend=2)
```
