# Embedding

In natural language processing, word embedding is a term used for the representation of words for the representation of words for text analysis, typically in the form of a real-valued vector that encodes the meaning of the word such that the workds that are closer in the vector space are expected to be similar in meaning

```python
from keras.models import Sequential
from keras.layers import Embedding
```

```python
docs = [
    'go india',
    'india india',
    'hip hip',
    'jeetega bhai jeetega india jeetega',
    'bharat mata ki jai',
    'kohli kohli',
    'sachin sachin',
    'dhoni dhoni',
    'modi ji ki jai',
    'inquilab zindabad',
]
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences, padding='post')
```

```python
sequences
```

```python
len(sequences)
```

```python
sequences.shape
```

# Embedding

```python
model = Sequential()
model.add(Embedding(17, output_dim=2, input_length=5)) # 17 - unique words in vocabulary, input_len is the word per list(shape[1] of sequences) 
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', metrics=['accuracy'])
```

```python
pred = model.predict(sequences)
print(pred)
```
