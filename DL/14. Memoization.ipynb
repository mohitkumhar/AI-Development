# Memoization
- Memoization is a technique that speeds up programs by storing and reusing results of function calls

```python
# fibonacci example

import time

def fib(n):
    if n == 1 or n == 0:
        return 1

    return fib(n - 1) + fib(n - 2)

start = time.time()
print("Fibonacci is: ", fib(36))
print("Total time taken: ", time.time() - start)
```

```python
# using memoization in this example to improve performance

def fib(n, record):
    if n in record:
        return record[n]

    record[n] = fib(n - 1, record) + fib(n - 2, record)
    return record[n]

start = time.time()
record = {0:1, 1:1}
print("fibonacci is: ", fib(36, record))
print("Total time taken: ", time.time() - start)
```
