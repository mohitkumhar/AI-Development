## What is Deep Learning?

Deep Learning is a subfield of Artificial Intelligence and Machine Learning that is inspired by the structure of a human brain.

Deep Learning algorithms attempt to draw similar conclusions as humans would by continually analyzing data with a given logical structure called Neural Network

### or

Deep Learning is part of broader family of <u>machine learning</u> methods based on <u>artificial neural networks</u> with <u>representation learning</u>.

Deep Learning Algorithms uses multiple layers to progressively extract higher-level features from the raw input. For Example, in <u>image processing</u>, lower layers may identify edges, while higher layers may identify the concepts relevent to a human such as digits or letters or faces.

###  Types of Neural Network

- ANN = Artifical Neural Network
- CNN = Convolutional Neural Network   (better for image)
- RNN = Recurrent Neural Network       (better for speech and text)
- GAN = Generative Adversarial Network (Better for text and images generation)

## Deep Learning VS Machine Learning
1. Data Dependency
2. Hardware Dependency
3. Training Time
4. Feature Selection
5. Interpretability


#### 1. Data Dependency:

Deeplearning is a data hungry
- Less Data ML Wins
- High Data DL Wins

#### 2. Hardware Dependency
- DL do the complex matrix multiplication, that is why it requires GPU
- Training is fast on GPU and slower on CPU

#### 3. Training Time
- Training Time is very high in DL (some time days or weeks)
- Training Time is low in ML (some minutes or hours)
- Prediction Time is fast in DL


#### 4. Feature Selection
- DL automatically extract features unlike ML where we provide feature to our ML algorithm




#### 5. Interpretability
- Interpretability in DL models means we dont have a clear reason for the output, like how the model arrived at a specific decision.
- It is often considered a "black box" due to its complex inner workings.

### Application Specific Integrated Circuit
1. TPU - Tensor Processing Unit - By Google (used in Google Colab)
2. Edge TPU - Used in Drone, watch, smart glass
3. NPU - Neural Processing Unit (accelerate work on mobile devices)

##### Scenario of using difference types of hardware
- For Small Projects   -  CPU
- For Large Projects   -  GPU, TPU
- For Smart Phone      -  Mobile CPU, GPU, DSP(Digital Signal Processor), NPU
- Smart Watch/ Glass   -  Edge TPU, NPU
