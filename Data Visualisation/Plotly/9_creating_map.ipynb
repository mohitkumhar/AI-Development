```python
import plotly.figure_factory as ff
import pandas as pd
```

```python
scope = ['New York']

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/minoritymajority.csv')

df.head(1)
```

```python
# df name
df_new = df[df['STNAME'].isin(scope)]

# list values
values = df_new['TOT_POP'].tolist()
fips = df_new['FIPS'].tolist()

```

```python
colorscale = ['red', 'blue', 'green', 'black']
```

```python
fig = ff.create_choropleth(
    fips=fips,
    values=values,
    scope=scope,
    colorscale=colorscale,
    round_legend_values=True,
    simplify_county=0,
    simplify_state=0,
    county_outline={
        'color':'rgb(15,15,55)',
        'width':0.5,
    },
    
    
)
```
