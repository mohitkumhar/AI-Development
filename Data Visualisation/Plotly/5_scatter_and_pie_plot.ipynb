```python
import chart_studio.plotly as py
from plotly import graph_objs as go
from plotly.offline import plot, iplot
import numpy as np
```

```python
n = 100
example_x = np.random.rand(n)
example_y = np.random.rand(n)
```

```python
trace = go.Scatter(
    x = example_x,
    y = example_y,
    mode = 'markers + lines'
)
```

```python
data = [trace]
iplot(data)
```

```python
# breaking each category
groups = ['Rent', 'Food', 'Bills', 'Miscellaneous']

# create amount
amount = [1000,518,331,277]

# style
colors = ['r', 'b', 'y', 'g']

trace = go.Pie(
    labels=groups,
    # Labels for each pie slice

    values=amount,
    # Values corresponding to the sizes of each pie slice

    hoverinfo='label+percent+value+text',
    # Information displayed on hover (label, percentage, value, text)

    textinfo='value',
    # Information displayed on the pie slices (only value)

    textfont=dict(size=18),
    # Font size for the text displayed on the pie slices

    marker=dict(
        colors=colors, 
        # Colors for each pie slice
        
        line=dict(
            color='black',
            # Color of the line around each pie slice
            
            width=2,
            # Width of the line around each pie slice
        )
    )
)

iplot([trace])
```
