```python
from plotly import graph_objs as go
import pandas as pd
from plotly.offline import iplot
```

```python
data = pd.read_csv('sample2.csv')

data.head()
```

```python
trace1 = go.Scatter(
    x=data.date,
    y=data.high,
    name='Tesla High',
    line=dict(
        color='#17BECF',
    ),
    opacity=0.8,
)

trace2= go.Scatter(
    x=data.date,
    y=data.low,
    name='Tesla Low',
    line=dict(
        color='#7F7F7F',
    ),
    opacity=0.8,
)


data = [trace1, trace2]

layout = dict(
    title = 'Tesla Stock Price - High Value'
)

fig = go.Figure(data=data, layout=layout)

iplot(fig)
```
