# Count Plot

```python
# Difference Between CountPlot and BarPlot
# -> CountPlot plots the count of the number of records by category
# -> BarPlot plots a value or metric for each category (by default, barplots plots the mean of a variable, by category)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset('tips')

sns.countplot(x="sex", data=data, hue='smoker', color='r', saturation=10)
# saturation=10: This parameter controls the saturation of the colors used in the plot. In this case, it sets the saturation to 10, making the color (specified by the color parameter) more intense or saturated.

plt.show()


```
