```python
# Bar Chart

import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd

data = sns.load_dataset('penguins')

order = ["Dream", "Torgersen", "Biscoe"]

sns.barplot(x = "island", y = "bill_length_mm", data=data, hue='sex',
           hue_order=["Female", "Male"], order=order, errorbar=('ci', 40), orient='v',
           saturation=100, errcolor='black', capsize=0.1, dodge=True, alpha=0.8) # for orient it horizontal both x and y should be number
# hue_order=["Female", "Male"]: This parameter specifies the order in which the 'hue' variable levels should appear.
# errorbar=('ci', 40): This parameter controls the error bars on the plot. 'ci' stands for confidence interval, and 40 represents the size of the confidence interval.
# orient='v': This parameter specifies the orientation of the plot. 'v' stands for vertical orientation.
# saturation=100: This parameter controls the saturation of the colors used in the plot. (maximum = 100)
# errcolor='black': This parameter sets the color of the error bars.
# capsize=0.1: This parameter sets the size of the caps at the ends of error bars.
# dodge=True: This parameter controls whether to separate the bars for different levels of the 'hue' variable.

sns.set(style="darkgrid")
# sns.set(style="darkgrid"): This function is used to set the aesthetic style of the plots created by Seaborn. Here, it sets the style to "darkgrid".


plt.show()
```
