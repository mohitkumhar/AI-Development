```python
# HeapMap Plot

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

data = np.linspace(1, 10, 20).reshape(4,5)  # This creates a 4x5 matrix of evenly spaced values from 1 to 10 using NumPy.

plot = sns.heatmap(data, vmin=0, vmax=12, cmap='PuOr', 
            annot=True, annot_kws={'fontsize':15, 'color':'r'},
            linewidths=4, linecolor='black',
             cbar=True, xticklabels=False, yticklabels=False,)
# vmin = 0: Sets the minimum value of the color scale to 0.
# vmax = 12: Sets the maximum value of the color scale to 12.
# cmap = 'PuOr': Sets the colormap to 'PuOr' (Purple-Orange).
# annot = True: Annotates each cell with its value.
# annot_kws = {'fontsize':15, 'color':'r'}: Sets the annotation's font size to 15 and color to red.
# linewidths = 4: Sets the width of the lines that divide the cells to 4.
# linecolor = 'black': Sets the color of the lines that divide the cells to black.
# cbar = True: Displays the color bar on the side.
# xticklabels = False: Hides the x-axis tick labels.
# yticklabels = False: Hides the y-axis tick labels.

plot.set(xlabel="Mohit", ylabel="Kumhar")
sns.set(font_scale=2)

plt.show()



```
