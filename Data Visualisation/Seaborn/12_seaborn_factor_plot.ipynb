# CatPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset('tips').head(20)

print(data.to_string())

sns.catplot(x='size', y='tip', data=data, hue='sex', 
            palette='Dark2', height=5, kind='point', )

# height=5: This parameter sets the height of the figure to 5 inches.

# kind='point': This parameter specifies the type of categorical plot to draw. In this case, it creates a point plot.

plt.show()
```
