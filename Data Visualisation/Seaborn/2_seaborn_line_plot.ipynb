# Line Plot

```python
# ploting with dataframe

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

x = pd.DataFrame({'var':var, 'var_1':var_1})  # Creating a pandas DataFrame named 'x' with two columns 'var' and 'var_1', using the lists 'var' and 'var_1'.

sns.lineplot(x='var', y='var_1', data=x)  # Creating a line plot using seaborn library where 'var' is plotted on the x-axis and 'var_1' on the y-axis using data from the DataFrame 'x'.

plt.show()
```

```python
# printing without dataframe

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

sns.lineplot(x=var, y=var_1)  # Ploting Graph Without DataFrame

plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

y = sns.load_dataset('penguins').head(20)  #  This function loads one of the built-in datasets provided by Seaborn named 'penguins'.
# This method is used to select the first 20 rows of the dataset loaded using .load_dataset().
     
a = sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=y, hue='sex', size=80,
                 style='sex', palette='Accent', markers=['o', '>'], dashes=True, legend=False,)
#  This parameter assigns the 'sex' column of the DataFrame 'y' to the 'hue' parameter, which is used to color the lines based on the values in the 'sex' column.
# This parameter sets the size of the markers used in the plot.
# This parameter assigns the 'sex' column of the DataFrame 'y' to the 'style' parameter, which is used to differentiate the style of markers based on the values in the 'sex' column.
# This parameter sets the color palette for the plot. Here, the 'Accent' palette is used.
# This parameter specifies the markers to be used for plotting. In this case, circles ('o') and triangles ('>') are used.

plt.title("Seaborn")

plt.show()


```
