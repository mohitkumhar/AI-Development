# Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset('tips').head(20)
sns.pairplot(data, vars=['total_bill', 'tip', 'size'], hue='sex', 
             hue_order=['Female', 'Male'], palette='Dark2',
             diag_kind='hist', markers=['<', 'o'])

# vars=['total_bill', 'tip', 'size']: Specifies which variables from the dataset should be included in the pair plot.

# hue_order=['Female', 'Male']: Specifies the order in which the genders should be displayed in the legend.

# diag_kind='hist': Specifies that histograms should be plotted on the diagonal subplots to visualize the distribution of each variable.

# markers=['<', 'o']: Sets the markers for the scatter plots, with '<' for females and 'o' for males, helping to differentiate between the genders visually.

plt.show()
print(data.to_string())




```
