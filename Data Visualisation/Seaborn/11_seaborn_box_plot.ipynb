# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset('tips')
# print(data)
sns.set(style='whitegrid')

sns.boxplot(x='day', y='total_bill', data=data, hue='time', 
            order=['Sun', 'Thur','Fri','Sat'],
            showmeans=True, 
            meanprops={'marker':'+', 'markeredgecolor':'r'},
            linewidth=1)

# showmeans=True: This parameter displays the mean value as a marker inside the box plot.

# meanprops={'marker':'+', 'markeredgecolor':'r'}: This parameter customizes the appearance of the marker representing the mean. It sets the marker style to '+' and the edge color to red ('r').

plt.show()
```

# Single Box Plot

```python
sns.set(style='whitegrid')

sns.boxplot(y='total_bill', data=data)

plt.show()
```
