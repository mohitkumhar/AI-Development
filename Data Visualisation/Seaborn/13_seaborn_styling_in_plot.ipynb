```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset('tips')

sns.set_style('darkgrid')

sns.set_context('paper', font_scale=1)
# sns.set_context('paper'): This function sets the context for the plot. 'paper' context is typically used for figures that will be included in academic papers. It adjusts the sizes of various plot elements to be appropriate for inclusion in a paper.
# font_scale=1: This parameter within set_context() adjusts the size of the fonts used in the plot. A font_scale of 1 means that the font size is not scaled relative to the base font size defined by the context.

sns.barplot(x ='day', y='total_bill', data=data, palette='Dark2')

plt.show()
```

# Removing Axes Lines and Figure Size

```python
sns.set_style('whitegrid')

plt.figure(figsize=(7,5))

sns.barplot(x ='day', y='total_bill', data=data)

sns.despine()

plt.show()
```
