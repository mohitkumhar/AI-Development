# Strip Plot

```python
# A Strip plot is very simple to understand. It is basically a scatter plot that differentiates different categories. Strip plots are considered a good alternative to a box plot or a violin plot.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset('tips')

sns.stripplot(x='day', y='total_bill', data=data, hue='sex', palette='Dark2', 
              linewidth=1, edgecolor='black',
              jitter=0.2, size=8, marker='o')

# jitter=0.2: Make x-axis (day) to prevent overlapping points, making it easier to see the distribution.

plt.show()

```
