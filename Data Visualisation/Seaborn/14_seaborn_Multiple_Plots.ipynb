# Facet Grid

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset('tips')

fg = sns.FacetGrid(data, col='day', hue='sex', palette='Dark2')
# col='day': Creates separate subplots for each unique value in the 'day' column.

fg.map(plt.bar, 'total_bill', 'tip', edgecolor='black').add_legend()  # This maps the data to a bar plot within each facet. It uses 'total_bill' as the x-axis and 'tip' as the y-axis.
# .add_legend(): Adds a legend to the plot, explaining the colors used for different sexes.

plt.show()

```
