# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset('penguins').head(20)

marker = {'Male': '<', 'Female': '*'}

sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=data, hue='sex', style='sex', size='sex',sizes=(100,20),palette='gist_ncar', alpha=0.8, markers=marker)
# style: Sets different markers for each category in 'sex' column (e.g., Male and Female)
# size: Adjusts the size of markers based on the values in the 'sex' column
# sizes: Defines the range of marker sizes; here, larger for female (100) and smaller for male (20)
# markers: Assigns specific markers to each category ('Male' and 'Female') based on the dictionary 'marker'

plt.plot()
```
