# Histogram Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = sns.load_dataset("penguins")

sns.displot(data["flipper_length_mm"], bins=[170,180,190,200,210,220,230], kde=True, rug=True, color='black', log_scale=False)
# bins: This parameter specifies the bins for the histogram in the distribution plot. Data points will be grouped into these bins.
# kde=True: This parameter specifies whether to plot the Kernel Density Estimate (KDE) curve along with the histogram. KDE is a smoothed, continuous version of the histogram.
# rug=True: This parameter specifies whether to draw small vertical lines (rug plot) along the x-axis to indicate the distribution of data points.
# log_scale=False: This parameter specifies whether to use a logarithmic scale on the y-axis. Here, it's set to False, so a linear scale is used.

plt.show()
```
