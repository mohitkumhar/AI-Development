# subplot()

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x,y,color='r')

x1 = [10,20,30,40]
plt.subplot(2,2,2)
plt.pie([1])

x1 = [10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x1)

x2 = ['a', 's', 'd', 'f']
y2 = [10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x1,x2)

plt.show()
```
