# Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random


x = np.random.randint(10, 60, 50)
print(x)
l = [10,20,30,40,50]


plt.hist(x, color = 'r', edgecolor='black',bins=l, range=(10, 60), cumulative=-1, bottom=10, align='mid', histtype='barstacked', orientation='vertical', rwidth=0.8, log=False, label="Mohit's Histo")

plt.title("HistoGram")
plt.xlabel("Python")
plt.ylabel("No.")

plt.axvline(30, color='black', linestyle='--', alpha=0.8)
plt.legend()

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

plt.plot()

```
