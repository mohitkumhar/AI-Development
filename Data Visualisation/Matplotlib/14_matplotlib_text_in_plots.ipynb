```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y, label='up')
# Plotting the graph with a label 'up'


plt.title("Mohit's Graph", fontsize=20)
plt.xlabel("Days", fontsize=15)
plt.ylabel("Python", fontsize=15)

plt.text(2,5, 'java', fontsize = 15, style='italic', bbox={'facecolor':'r'})
# Adding text 'java' at coordinates (2,5) with italic style and a red background box

plt.annotate("python", xy=(2,1), xytext=(3,3), arrowprops={'facecolor':'black', 'shrink':100})
# Annotating the graph with text 'python' at coordinates (2,1) pointing to (3,3) with an arrow


plt.legend(loc=2, facecolor='r', edgecolor='b', framealpha=0.4, shadow=True)
# Adding a legend at the top left (loc=2), with a red face color, blue edge color, partially transparent, and shadow

plt.show()
```
