# Linear Plot

```python
import matplotlib.pyplot as plt

x = [20,50,40,30,50,40,30,70]

plt.plot(x, color='r', marker='o', mec='g', alpha=0.6)
# mec gives a border to a marker

plt.show()
```

# Multiple Linear Plot

```python
import matplotlib.pyplot as plt

x1 = [20,50,40,30,50,40,30,70]
x2 = [40,20,10,40,30,60,80,40]

plt.plot(x1, color='r', linewidth=3, marker='D')
plt.plot(x2, color='m', linestyle='--', marker='o')

plt.show()
```
