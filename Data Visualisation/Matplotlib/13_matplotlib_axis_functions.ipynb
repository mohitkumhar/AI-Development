# xticks() & yticks()

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [3,1,5,2]

plt.plot(x, y)

plt.xticks(x, labels=['python', 'c', 'cpp', 'java']) # sets the x-axis
plt.yticks()


plt.show()
```

# xlim() & ylim()

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [3,1,5,2]

plt.plot(x, y)


plt.xlim(0,10)  # sets the limit of x-axis (from, to)
plt.ylim(0,10)  # sets the limit of y-axis (from, to)

plt.show()
```

# axis()

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [3,1,5,2]

plt.plot(x, y)

plt.axis([0,10, 0,7])  # art of writing xlim() and ylim() in one line
```
