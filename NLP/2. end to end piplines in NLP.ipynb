#### What is NLP Pipelines .?
NLP is a set of setps followed to build an end to end NLP software.
NLP software consists of the following steps
1. Data Acquisition
2. Text Preparation
   - Text Cleanup
   - Basic Preprocessing
   - Advance Preprocessing
3. Feature Engineering
4. Modeling
   - Model Building
   - Evaluation
5. Deployment
   - Deployment
   - Monitoring
   - Model Update

#### Points to Remember
- It is not Universal
- Deep learning Pipelines are slightly different
- Pipelines are non-linear
