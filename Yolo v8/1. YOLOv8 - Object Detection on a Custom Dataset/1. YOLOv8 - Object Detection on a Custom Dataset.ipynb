# 1. Detection

```python
! yolo task=detect mode=predict model=yolov8n.pt source='images/1.jpg'
```

# 2. Segmentation

```python
! yolo task=detect mode=predict model=yolov8n-seg.pt source='images/1.jpg'
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n-seg.pt') # Load a pretrained YOLOv8n segmentation model
model.predict(source='images/1.jpg') # predict on an image
```

# 3. Classification

```python
! yolo task=detect mode=predict model=yolov8n-cls.pt source='images/1.jpg'
```

```python
# Load a pretrained YOLOv8n classification model, train it on imagenette160 for 3 epochs and predict an image with it

from ultralytics import YOLO

model = YOLO('yolov8n-cls.pt')
model.train(data='imagenette160', epochs=3) # train the mode
model.predict(source='https://ultralytics.com/images/bus.jpg')
```
