# Chat Prompt Template in LangChain

A Chat Prompt Template in LangChain is a structure that defines how a conversation is started and managed using different types of messages. It helps guide the AI to respond in a specific way.

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate(
    [
        ('system', 'You are a help {domain} expert'),
        ('human', 'Explain in simple terms, what is {topic}')
    ]
)
```

```python
prompt = chat_template.invoke({'domain': 'cricket', 'topic': 'Dusra'})
```

```python
print(prompt)
```
