# <center>Open Source Model</center>

```python
!pip install langchain-huggingface
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
```

```python
HUGGINGFACE_ACCESS_TOKEN = "ACCESS_TOKEN"
```

```python
llm = HuggingFaceEndpoint(
    repo_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    task='text-generation',
    huggingfacehub_api_token=HUGGINGFACE_ACCESS_TOKEN,
    verbose=True,
    timeout=30
)
```

```python
model = ChatHuggingFace(llm=llm)
```

```python
result = model.invoke("What is thecapital of india")
```

### Using Hugging Face Model Without API, (local download)

```python
!pip install transformers
```

```python
# pytorch for cuda v11.8
!pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
```

```python
# [OPTIONAL] selecting the location of model downlaod
import os
os.environ['HF_HOME'] = r'D:/huggingface_cache'
```

```python
llm = HuggingFacePipeline.from_model_id(
    model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    task="text-generation",
    batch_size=1,
    device=-1,  # for CPU
    pipeline_kwargs=dict(
        temperature=0.5,
        max_new_tokens=100,
    )
)
```

```python
model = ChatHuggingFace(
    llm=llm,
    verbose=True,
)
```

```python
result = model.invoke("What is the capital of India")
```

```python
result
```
