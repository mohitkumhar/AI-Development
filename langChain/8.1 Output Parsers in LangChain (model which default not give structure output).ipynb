# Output Parsers

<b>Output Parsers</b> in LangChain help convert raw LLM responses into structured formats like JSON, CSV, Pydantic models, and more. They ensure consistency, validation, and ease of use in applications.

- There are many types of parsers, but mainly we study about these 4 here:
1. Str Output Parser
2. JSON Output Parser
3. Structured Output Parser
4. Pydantic Output Parser

**Note:** These Parsers can be run on any of the model, which may have default output parser or maynot
and also work on close as well as open source models
