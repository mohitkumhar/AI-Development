```python
from langchain_huggingface import HuggingFaceEmbeddings
from sklearn.metrics.pairwise import cosine_similarity
```

```python
repo_id = "sentence-transformers/all-MiniLM-L6-v2"
```

```python
embedding = HuggingFaceEmbeddings(model=repo_id, show_progress=True)
```

```python
documents = [
    "Virat kohli was the captian of team india",
    "Ms Dhoni gives the team india all icc trophy",
    "Jasprit is bowler of team india",
]
```

```python
query = "tell about ms dhoni"
```

```python
doc_embeddings = embedding.embed_documents(documents)
query_embeddings = embedding.embed_query(query)
```

```python
result = cosine_similarity([query_embeddings], doc_embeddings)
```

```python
result
```

```python
result = list(enumerate(result[0]))
```

```python
result
```

```python
index, conf = sorted(result, key=lambda x: x[1])[-1]
```

```python
print(documents[index])
print(f"Confidence: {conf}")
```
