1. **Bagging (Bootstrap Aggregating)**:
   - **How it works**: It creates many decision trees by training each on random parts of the data.
   - **Feature use**: Each tree looks at all the features (columns) when making splits.
   - **Final decision**: For predictions, it averages the results (for numbers) or takes the most common answer (for categories).
   - **Purpose**: It helps to reduce mistakes and prevent the model from learning too much from just the training data.

2. **Random Forest**:
   - **How it works**: It’s similar to bagging but adds more randomness. When making splits, each tree only looks at a few features instead of all of them.
   - **Feature use**: Each tree uses a different set of features at each step, which makes the trees less alike.
   - **Final decision**: Like bagging, it averages or takes the most common answer.
   - **Purpose**: This extra randomness makes the model better at handling new data and avoids overfitting.

In short, **Random Forest adds more randomness by picking different features**, while **Bagging uses all the features** for each tree.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_features=5, n_redundant=0, n_informative=5, n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X, columns=['col1', 'col2', 'col3', 'col4', 'col5'])
df['target'] = y
print(df.shape)
df.head()
```

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5], df.iloc[:, -1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf = RandomForestClassifier(max_features=2)

rf.fit(df.iloc[:,:5], df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
```
