```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('BostonHousing.csv')
```

```python
df.head()
```

```python
X = df.drop(columns=['medv'])
y = df['medv']
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [('lr', lr), ('dt', dt), ('svr', svr)]
```

```python
for estimator in estimators:
    scores = cross_val_score(estimator[1], X, y, scoring='r2', cv=10)
    print(estimator[0], np.round(np.mean(scores), 2))
```

```python
from sklearn.datasets import fetch_california_housing
import numpy as np

# Load the dataset
data = fetch_california_housing()

# Separate features and target
X, y = data['data'], data['target']

# Check the shape of features and target
print(X.shape)  # (20640, 8)
print(y.shape)  # (20640,)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [('lr',lr),('dt',dt),('svr',svr)]
```

```python
for estimator in estimators:
  scores = cross_val_score(estimator[1],X,y,scoring='r2',cv=10)
  print(estimator[0],np.round(np.mean(scores),2))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr, X, y, scoring='r2', cv=10)
print("Voting Regressor: ", np.round(np.mean(scores), 2))
```

```python
for i in range(1, 4):
    for j in range(1, 4):
        for k in range(1, 4):
            vr = VotingRegressor(estimators, weights=[i, j, k])
            scores = cross_val_score(vr, X, y, scoring='r2', cv=10)
            print('for i={}, j={}, k={}'.format(i, j, k), np.round(np.mean(scores), 2))
            
```

```python
dt1 = DecisionTreeRegressor(max_depth=1)
dt2 = DecisionTreeRegressor(max_depth=2)
dt3 = DecisionTreeRegressor(max_depth=3)
dt4 = DecisionTreeRegressor(max_depth=4)
dt5 = DecisionTreeRegressor(max_depth=5)
dt6 = DecisionTreeRegressor(max_depth=None)
```

```python
estimators = [('dt1', dt1), ('dt2', dt2), ('dt3', dt3), ('dt4', dt4), ('dt5', dt5), ('dt6', dt6)]
```

```python
for estimator in estimators:
    scores = cross_val_score(estimator[1], X, y, scoring='r2', cv=10)
    print(estimator[0], np.round(np.mean(scores), 2))
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr, X, y, scoring='r2', cv=10)
print('Voting Regressor: ', np.round(np.mean(scores), 2))
```
