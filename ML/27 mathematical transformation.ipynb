### Log Transform
- **Purpose:** Reduces skewness and handles outliers.
- **How It Works:** Applies the natural logarithm (or base 10/2) to each data point.
- **Use Cases:** Positive data; often used in finance and biology.

### Reciprocal Transform
- **Purpose:** Reduces skewness and handles outliers, particularly for large values.
- **How It Works:** Takes the reciprocal (1/x) of each data point.
- **Use Cases:** Positive data; not suitable if data contains zeros or negative values.

### Square Root Transform
- **Purpose:** Reduces skewness and spreads out smaller values.
- **How It Works:** Takes the square root (âˆšx) of each data point.
- **Use Cases:** Non-negative data; often used in count data and measurements.

Each of these transformations helps make data more normal (Gaussian), which is useful for various statistical analyses and machine learning models.
