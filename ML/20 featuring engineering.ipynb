# 2. Feature Construction

Feature construction is the process of creating new features from existing data to improve the performance of machine learning models. For example, generating a "Total Price" feature by multiplying "Quantity" and "Price per Unit" in a sales dataset.

# 3. Feature Scaling

Feature scaling is the process of making sure all features in a dataset have similar ranges. For example, transforming features to a range of [0, 1] using Min-Max scaling.

# 4. Featuring Extraction

Feature extraction is the process of selecting and transforming raw data into a set of features that better represent the data for a specific task. For example, extracting edges from an image to use in image recognition.
