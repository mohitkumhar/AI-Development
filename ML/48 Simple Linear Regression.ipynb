```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement1.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'], df['package']) # owr main observation is to just check if the data is linearly alligned
plt.xlabel('CGPA')
plt.ylabel('Package (in LPA)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test)
```

```python
# what patterns does this algo see while training
plt.scatter(df['cgpa'], df['package'])
plt.plot(X_train, lr.predict(X_train), color='red')
plt.xlabel('CGPA')
plt.ylabel('Package (in LPA)')
```

```python
# formula used: y = mx + c
lr.coef_ # this is a slope of the line, which is used by model for prediction
```

```python
lr.intercept_ # this the value of 'c' as the constant
```
