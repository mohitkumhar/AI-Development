```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('student_clustering.csv')
df.shape
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'], df['iq'])
```

```python
# using elbo method
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

```python
y_means
```

```python
X[y_means == 1] # first cluster 
```

```python
plt.scatter(X[y_means == 0,0], X[y_means == 0,1], color='blue')
plt.scatter(X[y_means == 1,0], X[y_means == 1,1], color='red')
plt.scatter(X[y_means == 2,0], X[y_means == 2,1], color='green')
plt.scatter(X[y_means == 3,0], X[y_means == 3,1], color='yellow')
```

Clustering works same for all dimention of data 

## 3-D Clustering

```python
from sklearn.datasets import make_blobs

centroids = [(-5,-5,5), (5,5,-5), (3.5,-2.5,4), (-2.5,2.5,-4)]
cluster_std = [1,1,1,1]

X, y = make_blobs(n_samples=200, cluster_std=cluster_std, centers=centroids, n_features=3, random_state=1)
```

```python
X
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming X is your data array with 3 columns
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], X[:, 2])

# Add axis labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```

```python
wcss = []

for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(X)
```

```python
df = pd.DataFrame()

df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]
df['label'] = y_pred
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np

# Example dataframe
# df = pd.DataFrame({'col1': ..., 'col2': ..., 'col3': ..., 'label': ...})

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot data points with color based on 'label'
sc = ax.scatter(df['col1'], df['col2'], df['col3'], c=df['label'], cmap='viridis')

# Add axis labels
ax.set_xlabel('col1')
ax.set_ylabel('col2')
ax.set_zlabel('col3')

# Add color bar
plt.colorbar(sc)

# Show plot
plt.show()
```
