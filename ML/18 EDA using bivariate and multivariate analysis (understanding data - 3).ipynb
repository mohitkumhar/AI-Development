# 1. Scatterplot (Numerical - Numerical)

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='time')
```

# 2. Barplot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic, hue='Sex')
```

# 3. Boxplot (Numerical - Categorial)

```python
sns.boxplot(x='Sex', y='Age', data=titanic, hue='Survived')
```

# 4. Distplot (Numerical - Categorical)

```python
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'])
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'])
```

```python
titanic.head()
```

# Heatmap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
# also be done using pandas
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived'], ), annot=True)
```
