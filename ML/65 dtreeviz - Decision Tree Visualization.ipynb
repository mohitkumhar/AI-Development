```python
from sklearn.datasets import *
from dtreeviz.trees import *

from IPython.display import Image, display_svg, SVG
```

```python
clas = tree.DecisionTreeClassifier(max_depth=2)
iris = load_iris()
```

```python
X_train = iris.data
y_train = iris.target

clas.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clas) # this is default visualisation, provided by sklearn tree
```

## Visualization using dtreeviz

### 1. Classification

```python
viz = dtreeviz(clas, 
               X_train,
               y_train,
               feature_names=iris.feature_names, 
               class_names=["setosa", "versicolor", "virginica"])
viz
```

```python


viz = dtreeviz(clas, 
               X_train,
               y_train,
               target_name='price',
               feature_names=iris.feature_names, 
               class_names=["setosa", "versicolor", "virginica"], 
               scale = 1.5,
               orientation='LR')
viz
```

```python
from sklearn import tree
from sklearn.datasets import load_iris
from dtreeviz import model
from IPython.display import SVG

# Load dataset and train classifier
iris = load_iris()
X_train = iris.data
y_train = iris.target

clas = tree.DecisionTreeClassifier()  
clas.fit(X_train, y_train)

# Plot tree using sklearn's plot_tree
from sklearn.tree import plot_tree
plot_tree(clas)

# Visualize tree using the new dtreeviz API
viz_model = model(clas, 
                  X_train,
                  y_train,
                  feature_names=iris.feature_names, 
                  class_names=list(iris.target_names))  # Fix: class names must be a list

# Render the visualization
viz_model.view()
```
