# <center>Machine Learning</center>

Machine Learning is a field of computer science that uses statistical techniques to give computer systems the ability to <b><u>Learn with data</u></b>, without being explicitly programmed

<pre>
&rarr; Traditional Programming
              +------------+
Data     ---> |            |
Program  ---> |  Computer  |  ---> Output
              |            |
              +------------+
<br><br>
&rarr; Machine Learning
              +------------+
Data     ---> |            |
Output   ---> |  Computer  |  ---> Program
              |            |
              +------------+
</pre>


<h3>Data Mining: </h3>Data mining is the process of extracting and discovering patterns in large data sets using techniques from machine learning, statistics, and database systems.
