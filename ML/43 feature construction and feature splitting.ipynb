## 1. Feature Construction
Feature Construction (also known as feature engineering) is the process of creating new input features from raw data that can help improve the performance of machine learning models. The goal is to transform or derive new variables that represent the underlying structure of the data in a way that makes it easier for models to learn meaningful patterns.

## 2. Feature Splitting
Feature Splitting refers to breaking down a single feature into multiple components, often to better capture information or relationships that might be obscured in a combined format.

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df = pd.read_csv('train.csv')[['Age', 'Pclass', 'SibSp', 'Parch', 'Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
np.mean(cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=20))
```

## Applying Feature Construction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1  # adding 1 to count that passenger 
```

```python
X.head()
```

```python
def myFunc(num):
    if num == 1:
        # alone
        return 0
        
    elif num > 1 and  num < 5:
        # small family
        return 1
        
    else:
        # large family
        return 2
```

```python
myFunc(4)
```

```python
X['Family_type'] = X['Family_size'].apply(myFunc)
```

```python
X.head()
```

```python
X.drop(columns=['SibSp', 'Parch', 'Family_size'], inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=20))
```

## Feature Splitting


```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df['Name']
```

```python
# extraction the title of the person (like Mr, Miss, Mrs)  from name column

df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
```

```python
df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
```

```python
df[['Title', 'Name']]
```

```python
# errors='coerce' argument means that if there are any non-numeric values in the column you're trying to convert, those values will be replaced with NaN
df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce')  # Convert 'Survived' column to numeric
(df.groupby('Title')['Survived'].mean()).sort_values(ascending=False)
```
