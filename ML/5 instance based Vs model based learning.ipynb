# Instance-Based Learning

Instance-based learning, also called memory-based learning, is a type of machine learning. Here, the model stores examples from the training data to make predictions. It does not create a new model from the data but uses the stored examples to predict directly.

## Key Points:

- **Storage of Instances:** The model keeps the training examples in memory.
- **Prediction:** When new data comes in, the model compares it to stored examples and predicts based on similarity.
- **Example Algorithm:** K-Nearest Neighbors (KNN)

# Model-Based Learning

Model-based learning is a type of machine learning where the model learns a function or a set of rules from the training data. The model builds an abstract representation of the data and uses this to make predictions.

## Key Points:

- **Learning a Model:** The model creates a function or set of rules based on the training data.
- **Prediction:** When new data comes in, the model uses the learned function or rules to make predictions.
- **Example Algorithms:** Linear Regression, Decision Trees, Neural Networks
