```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
import os
```

```python
load_dotenv()
```

```python
os.environ["LANGCHAIN_PROJECT"] = "Sequencial LLM App"  # change the name of project 
```

```python
prompt1 = PromptTemplate(
    template='Generate a detailed report on {topic}',
    input_variables=['topic']
)

prompt2 = PromptTemplate(
    template='Generate a 5 pointer summary from the following text \n {text}',
    input_variables=['text']
)
```

```python
model = ChatOpenAI(
    base_url="https://openrouter.ai/api/v1",
    temperature=0.5
)
```

```python
# add configuration of langsmith
# 1. like adding metadata, tags and chage the name of the trace

config = {
    "run_name": "Sequential Chain Trace",
    "tags": ["llm app", "report generation", "summarization"],
    "metdata": {"model1": "gpt-40-mini", "model_temp": 0.5, "parser": "stroutputparser"},
}
```

```python

parser = StrOutputParser()

chain = prompt1 | model | parser | prompt2 | model | parser

result = chain.invoke({'topic': 'Unemployment in India'})

```
